```python
# IMPORT LIBRARIES 
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# IMPORT DATASET 
training_set = pd.read_csv('C:/Users/asperumal/Desktop/ML Classification Package/3. Logisitic Regression/Train_Titanic.csv')
```

```python
#VIEW DATASET (FIRST 10 VALUES IN THE DATASET)
training_set.head(10)
```

```python
#VEIW DATASET (LAST 10 VALUES IN THE DATASET )
training_set.tail(10)
```

```python
#EXPLORING THE DATA / VISUALISING THE DATA 
survived = training_set[training_set['Survived'] == 1]
not_survived = training_set[training_set['Survived'] == 0]
```

```python
#PRINT THE TOTAL NO OF PASSENGERS #EXPLORING THE DATA / VISUALISING THE DATA 
print ('Total' , len(training_set))
print('Total survived ' , len(survived))
print ('Total not survived' , len(not_survived))
```

```python
#PRINT THE PERCENTAGE OF PEOPLE WHO SURVIVED #EXPLORING THE DATA / VISUALISING THE DATA 
print ('% survived' , 1.*len(survived)/len(training_set)*100 )
print('% not survived', 1. *len(not_survived)/len(training_set)*100)
```

```python
# EXPLORING THE DATA / PLOTING THE DIFFERENT CLASS 
plt.figure (figsize = [10, 15])
plt.subplot(211)
sns.countplot(x = 'Pclass', data = training_set)
plt.subplot(212)
sns.countplot(x = 'Pclass', hue = 'Survived' , data = training_set)
plt.show()
```
